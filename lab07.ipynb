## Zadanie 2

Zadanie należy wykonać wykorzystując program `BagOfWords` z poprzednich zajęć.


Przerób ją tak, żeby interpunkcja, cyfry i wszelkie inne znaki nie przeszkadzały w parsowaniu tekstu. Uruchom ja na [tekście hamleta](http://www.gutenberg.org/cache/epub/1787/pg1787.txt). Ile razy występuje słowo hamlet? Jak brzmi 10 najczęściej występujących słów?

```python
import re
import requests
from io import TextIOWrapper
from collections import defaultdict

from lab07.ex2 import BagOfWords


def init(self, arg) -> None:
    self._data = defaultdict(int)

    if isinstance(arg, TextIOWrapper):
        with arg as f:
            line = re.sub(r'[^a-zA-Z]', ' ', f.readline()).lower()
            while line:
                for word in line.split():
                    self._data[word] += 1
                line = re.sub(r'[^a-zA-Z]', ' ', f.readline()).lower()
    
    if isinstance(arg, str):
        arg = re.sub(r'[^A-Za-z]', ' ', arg).lower()
        for word in arg.split():
            self._data[word] += 1

BagOfWords.__init__ = init

bow = BagOfWords(requests.get("https://www.gutenberg.org/cache/epub/1787/pg1787.txt").text)

print("*" * 15, "  Hamlet occurencies   ", "*" * 15)
print("hamlet", bow["hamlet"])

print("*" * 15, "Ten most frequent words", "*" * 15)
for i, word in enumerate(bow):
    print(word, bow[word])
    if i == 9:
        break
```

## Zadanie 3

Wykorzystując `pickle` zapisz i odczytaj klasy z poprzedniego zadania nakarmonej Hamletem. Porównaj metody i rozmiar.

```python
import sys
import pickle


with open("hamlet.pkl", "wb") as f:
    pickle.dump(bow, f)

with open("hamlet.pkl", "rb") as f:
    test_bow = pickle.load(f)

BagOfWords.__eq__ = lambda self, other: self._data == other._data

print(bow == test_bow)
print(set(dir(bow)) == set(dir(test_bow)))
print(sys.getsizeof(bow) == sys.getsizeof(test_bow))
```

## Zadanie 4

Wykorzystując [https://gist.github.com/pamelafox/986163](https://gist.github.com/pamelafox/986163) podaj aktualną godzinę we:
- wszystkich krajach, wyświetlając je zgrupowane względem kontynentów,
- przeprowadź symulacyjne wyświetlanie kolejnych krajów w miarę jak w danym kraju wybija północ, opóżnienie wyświetlania ustaw proporcjonalne do realnego czasu

```python
import time
from datetime import datetime
from collections import defaultdict

import pytz

from lab07.ex4 import countries

get_time_in_timezone = lambda timezone: datetime.now(pytz.timezone(timezone)).strftime("%H:%M:%S")
```

```python
continents = defaultdict(list)
for c in countries:
    continents[c["continent"]].append((c["name"], get_time_in_timezone(c["timezones"][0])))

for continent in continents:
    print(continent)
    for country in continents[continent]:
        print("  ", country[0], country[1])
```

```python
timezones = defaultdict(list)
for c in countries:
    timezones[get_time_in_timezone(c["timezones"][0])].append(c["name"])

for k, v in sorted(timezones.items(), key=lambda x: x[0], reverse=True):
    print("Midnight")
    for country in v:
        print("  ", country)
    time.sleep(1)
```

## Zadanie 5

Dla klasy `BagOfWords` napisz metody `save` oraz `load` wykorzystujące `json`'a do zapisu i odczytu danych.

```python
import json


def save_json(self, filename):
    with open(filename, "w") as f:
        json.dump(self._data, f)

def load_json(self, filename):
    with open(filename, "r") as f:
        self._data = json.load(f)

BagOfWords.save = save_json
BagOfWords.load = load_json

bow.save("hamlet.json")

test_bow = BagOfWords("")
test_bow.load("hamlet.json")

print(bow == test_bow)
```
